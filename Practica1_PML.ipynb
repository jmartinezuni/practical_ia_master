```python
def calculate_sums(n, array):
    """Calculates the sums of all subsets of a given array."""
    sums = set()  # Use a set to avoid duplicates
    
    # Generate all possible subsets using binary numbers
    for i in range(1, 2 ** n):  # Start at 1 to exclude the empty subset
        current_sum = 0
        for j in range(n):
            # Check if the j-th bit is set
            if (i & (1 << j)) != 0:
                current_sum += array[j]
        sums.add(current_sum)
    
    return sorted(sums)  # Sort the sums in ascending order

# Example usage
print("Enter an integer less than or equal to 10")
n = int(input())
print(f"Now enter the {n} elements of the array one by one")
array = []
for i in range(n):
    print(f"Element {i + 1}:")
    array.append(int(input()))

# Check if the input meets the specified conditions
if len(array) == n and all(1 <= x <= 99 for x in array) and len(set(array)) == n and n <= 10:
    sums = calculate_sums(n, array)
    print(f"The possible sums are: {sums}")
else:
    print("The array must meet the specified conditions.")
```

```python
def dif_simetrica(a,b):
    len_a=len(a)
    len_b=len(b)
    arreglo=[]
    diccionario_ind1= {i+1: valor for i, valor in enumerate(sorted(a))}
    diccionario_ind2= {i+1: valor for i, valor in enumerate(sorted(b))}
    for i in range(min(len_a,len_b)):
        if diccionario_ind1[i+1]!=diccionario_ind2[i+1]:
            item1=(diccionario_ind1[i+1],i+1)
            item2=(diccionario_ind2[i+1],i+1)
            arreglo.append(item1)
            arreglo.append(item2)
    for i in range(min(len_a,len_b),max(len_a,len_b)):
        if i+1 in diccionario_ind1:
            item1=(diccionario_ind1[i+1],i+1)
            arreglo.append(item1)
        if i+1 in diccionario_ind2:
            item2=(diccionario_ind2[i+1],i+1)
            arreglo.append(item2)
    return arreglo

a={1,3,5,7}
b={1,2,4,5}

print("sean los conjuntos a: "+str(a)+" y b: "+str(b))
print("la diferencia simetrica de los conjuntos indexados es:"+str(dif_simetrica(a,b)))
print("y la distancia es: "+str(len(dif_simetrica(a,b))))
```

```python


def numeropalabrasconvocalesdist(texto,n):
    palabras=texto.split()
    contador=0
    arreglo=[]
    for palabra in palabras:
        conjunto=set()
        for caracter in palabra:
            if caracter in ('a','e','i','o','u'):
                conjunto.add(caracter)
        if len(conjunto)>=n:
            contador=contador+1
            arreglo.append(palabra)
    return (contador,arreglo)

def calculo_n(texto):
    palabras=texto.split()
    suma=0
    for palabra in palabras:
        try:
            valor_entero = int(palabra)
            suma=suma+valor_entero
        except ValueError:
            suma=suma
    residuo=suma%5
    n=residuo+1
    return (suma,residuo,n)

print("escriba el texto a analizar:")
texto=input()
suma,residuo,n=calculo_n(texto)

contador,arreglo=numeropalabrasconvocalesdist(texto,n)
print(f"la suma de los numeros es: {suma}, el residuo es: {residuo}, y el n es: {n}" )
print(f"El número de las palabras con n vocales es {contador}, y las palabras son: {str(arreglo)}")
```

```python

def valida_bashuska(arreglo):
    
    if len(arreglo)%2==1 | arreglo[0]!=-1*arreglo[1] | arreglo[0]>=0:
        #print("falso por inicio")
        return False
    if len(arreglo)==2:
        return True
    tamaño=arreglo[-1]
    tamaño_restante=tamaño
    arr_hijos=arreglo[1:-1]
    if arr_hijos[0]>=0 | arr_hijos[-1]<=0 | ((-1*arr_hijos[0] + arr_hijos[-1])>=tamaño):
        #print("falso por hijos")
        return False
    arreglo_final=arr_hijos
    condicion=True
    while tamaño_restante>0:
        try:
            indice=arreglo_final.index(-1*arreglo_final[0])
            condicion=condicion & valida_bashuska(arreglo_final[0:indice+1])
            
            tamaño_restante=tamaño_restante-arreglo_final[indice]
            if tamaño_restante<=0:
                #print(" el tamaño restante es:"+str(tamaño_restante)+" analizando el arreglo "+str(arreglo_final))
                #print("falso por tamaño restante")
                return False
            if indice==len(arreglo_final)-1:
                return condicion
            arreglo_final=arreglo_final[indice+1:]
        except:
            #print(" el tamaño restante es:"+str(tamaño_restante)+" analizando el arreglo "+str(arreglo_final))
        
            #print("falso por except")
            return False
    return condicion
    

print("escribe la cantidad de lineas a leer")
n=int(input())
arreglo_lineas=[]
for i in range(n):
    print(f"escriba la linea {i+1}:")
    linea=input()
    arreglo_lineas.append(linea)

for linea in arreglo_lineas:
    valores=linea.split()
    arreglo_valores=[]
    for valor in valores:
        try:
            valor_entero = int(valor)
            arreglo_valores.append(valor_entero)
        except ValueError:
            print("error en los valores, deben ser enteros")
    if valida_bashuska(arreglo_valores):
        resultado=":-) babushkas!"
    else:
        resultado=":-( intenta de nuevo."
    print(resultado)
```
